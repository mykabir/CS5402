```python
# Importing necessary packages
from surprise import SVD
from surprise import NMF
from surprise import Dataset
from surprise import evaluate, print_perf
import os
from surprise import Reader
from surprise import KNNBasic
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib 
%matplotlib inline
```

```python
#load data from a file 
file_path = os.path.expanduser('restaurant_ratings.txt')
reader = Reader(line_format='user item rating timestamp', sep='\t')
```

```python
# Question 5. SVD
data = Dataset.load_from_file(file_path, reader=reader)
data.split(n_folds=3)
algo = SVD()
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf)
```

```python
# Question 6. PMF
data = Dataset.load_from_file(file_path, reader=reader)
data.split(n_folds=3)
algo = SVD(biased=False) #PMF
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf)
```

```python
# Question 7. NMF
data = Dataset.load_from_file(file_path, reader=reader)
data.split(n_folds=3)
algo = NMF()
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf)
```

```python
# Question 8. User based collaborative filtering
data = Dataset.load_from_file(file_path, reader=reader)
data.split(n_folds=3)
algo = KNNBasic(sim_options = {
        'user_based': True 
        })
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf)
```

```python
# Question 9. Item based collaborative filtering
data = Dataset.load_from_file(file_path, reader=reader)
data.split(n_folds=3)
algo = KNNBasic(sim_options = {
                'user_based': False 
                })
perf = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf)
```

### Question 10
#### Using same data for all the algos

```python
data = Dataset.load_from_file(file_path, reader=reader)
data.split(n_folds=3)
```

```python
# SVD
algo = SVD()
perf_SVD = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf_SVD)

#PMF
algo = SVD(biased=False) #PMF
perf_PMF = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf_PMF)

#NMF
algo = NMF()
perf_NMF = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf_NMF)

#User-based collaborative filtering
algo = KNNBasic(sim_options = {
        'user_based': True 
        })
perf_UserBased = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf_UserBased)


#Item-based collaborative filtering
algo = KNNBasic(sim_options = {
                'user_based': False 
                })
perf_ItemBased = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf_ItemBased)
```

```python
# Question 10-13. Printing the performance evaluation for all 3 folds and mean
print_perf(perf_UserBased)
print_perf(perf_ItemBased)
print_perf(perf_SVD)
print_perf(perf_PMF)
print_perf(perf_NMF)
```

```python
# Question 14
#User-based collaborative filtering. MSD
algo = KNNBasic(sim_options = {
        'name': 'MSD',
        'user_based': True 
        })
perf_UserBased_MSD = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf_UserBased_MSD)

#User-based collaborative filtering. Cosine
algo = KNNBasic(sim_options = {
        'name': 'cosine',
        'user_based': True 
        })
perf_UserBased_cosine = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf_UserBased_cosine)

#User-based collaborative filtering. Pearson.
algo = KNNBasic(sim_options = {
        'name': 'pearson',
        'user_based': True 
        })
perf_UserBased_pearson = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf_UserBased_pearson)
```

```python
# Question 14
# Item-based collaborative filtering. MSD. 
algo = KNNBasic(sim_options = {
        'name': 'MSD',
        'user_based': False 
        })
perf_ItemBased_MSD = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf_ItemBased_MSD)

# Item-based collaborative filtering. Cosine. 
algo = KNNBasic(sim_options = {
        'name': 'cosine',
        'user_based': False 
        })
perf_ItemBased_cosine = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf_ItemBased_cosine)

# Item-based collaborative filtering. Pearson. 
algo = KNNBasic(sim_options = {
        'name': 'pearson',
        'user_based': False 
        })
perf_ItemBased_pearson = evaluate(algo, data, measures=['RMSE', 'MAE'])
print_perf(perf_ItemBased_pearson)
```

```python
#Question 15 
#Finding best K for User Based collaborative filtering
user_Based_RMSE_of_different_k = []
user_Based_MAE_of_different_k = []
for i in range(1,30):
    algo = KNNBasic(k=i, sim_options = {
            'name': 'MSD',
            'user_based': True 
            })
    perf_UserBased_MSD = evaluate(algo, data, measures=['RMSE', 'MAE'], verbose=0)
    print("________________________________")
    print("K= ", i)
    print_perf(perf_UserBased_MSD)
    user_Based_MAE_of_different_k.append(np.mean(perf_UserBased_MSD['mae']))
    user_Based_RMSE_of_different_k.append(np.mean(perf_UserBased_MSD['rmse']))
```

```python
# Question 15 
# Finding best K for Item Based collaborative filtering
item_Based_RMSE_of_different_k = []
item_Based_MAE_of_different_k = []
for i in range(1,30):
    algo = KNNBasic(k=i, sim_options = {
            'name': 'MSD',
            'user_based': False 
            })
    perf_ItemBased_MSD = evaluate(algo, data, measures=['RMSE', 'MAE'], verbose=0)
    print("________________________________")
    print("K= ", i)
    print_perf(perf_ItemBased_MSD)
    item_Based_MAE_of_different_k.append(np.mean(perf_ItemBased_MSD['mae']))
    item_Based_RMSE_of_different_k.append(np.mean(perf_ItemBased_MSD['rmse']))
```

```python
#Best RMSE for User Based Collaborative filtering
min_RMSE_index = np.argmin(user_Based_RMSE_of_different_k)
print("Best K= ", min_RMSE_index)
print("Best RMSE= ", np.min(user_Based_RMSE_of_different_k))
```

```python
# Rmse Chart for User Based Collaborative filtering
series = pd.Series(name='rmse', data=user_Based_RMSE_of_different_k)
ax = sns.barplot(series.index, series.values)
ax.set_ylabel('RMSE values')
ax.set_xlabel('Number of K')
ax.set_xlim(1, 30)
ax.set_ylim(.7, 1.2)
```

```python
#Best RMSE for User Based Collaborative filtering
min_RMSE_index = np.argmin(item_Based_RMSE_of_different_k)
print("Best K= ", min_RMSE_index)
print("Best RMSE= ", np.min(item_Based_RMSE_of_different_k))
```
